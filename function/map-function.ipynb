```python
def square(x):
    return x*x

square(5)
```

```python
numbers=[1,2,3,4,5,6]

list(map(square, numbers))
```

```python
## lanbda function with map

list(map(lambda x: x * x, numbers))
```

```python
## Map Multiple Iterables

numbers1=[1,2,3]
numbers2=[4,5,6]

added_numbers=list(map(lambda x,y:x+y, numbers1, numbers2))
print(added_numbers)
```

```python
## Using Map to Convert String Lists to Integers

str_numbers=['1','2','3','4','5','6']
int_nums=list(map(int, str_numbers))
print(str_numbers)
print(int_nums)
```

```python
words=['apple', 'banana', 'cherry']
upper_words=list(map(str.upper, words))
print(words)
print(upper_words)
```

```python
def get_name(person):
    return person.get('name')

people=[
    {'name':"Nishit", 'age':28},
    {'name':"Nalin", 'age':29},
    {'name':"Srivastava", 'age':30},
]

list(map(get_name, people))
```
