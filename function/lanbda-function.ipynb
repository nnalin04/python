```python
def add(a, b):
    return a + b
print(add(5, 3))
```

```python
addition=lambda x, y: x + y

print(addition)
print(addition(5, 3))
```

```python
def even_or_odd(num):
    return num % 2 == 0

print(even_or_odd(4))
print(even_or_odd(3))
```

```python
even_or_odd = lambda num: num % 2 == 0

print(even_or_odd(4))
print(even_or_odd(3))
```

```python
numbers=[1,2,3,4,5,6]
def suqare(number):
    return number ** 2

suqare(2)
```
