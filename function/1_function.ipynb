```python
# Variable Length Arguments
# positional arguments

def print_numbers(*args):
    for num in args:
        print(num)
```

```python
print_numbers(1,2,3,4,5,6,7, "Nishit")
```

```python
# Keywords arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key} : {value}")
```

```python
print_details(name="Nishit",age=29,country="India")
```

```python
def print_details(*args, **kwargs,):
    for val in args:
        print(f"Positional argument: {val}")
    for key,value in kwargs.items():
        print(f"{key} : {value}")
```

```python
print_details(1,2,3,4,name="Nishit",age=29,country="India")
```
