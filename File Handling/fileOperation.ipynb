```python
### Copying Content from One File to Another
## To read content from a source file and write it to a destination file:

with open('example.txt', 'r') as source_file:
    content=source_file.read()

with open('destination.txt', 'w') as destination_file:
    destination_file.write(content)
```

```python
## Count Lines, Words, and Characters in a Text File
def count_file_stats(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len((line.split())) for line in lines)
        char_count = sum(sum(len((word)) for word in line.split()) for line in lines)
    return line_count, word_count, char_count
lines, words, char= count_file_stats('example.txt')
print(f'lines: {lines}, words: {words}, char: {char}')

 
```

```python
### Reading and Writing Using w+ Mode
## The w+ mode opens a file for both reading and writing. If the file does not exist, it is created. If it exists, its content is truncated (overwritten).

with open('example.txt', 'w+') as file:
    file.write('Hello World\n')
    file.write('THis is anew line.\n')
    
    ## move the file cursore to the begining
    file.seek(0)

    ## read the content of the file
    content=file.read()
    print(content)
```
