```python
import os

cwd=os.getcwd()
print(cwd)
```

```python
## create new diratory

new_drictory="package"
os.mkdir(new_drictory)
print(f"Drictory '{new_drictory}' created")
```

```python
## Listing files and drictories

items=os.listdir('.')
print(items)
```

```python
### Join paths
dir_name="folder"
file_name="file.txt"
full_path=os.path.join(os.getcwd(), dir_name, file_name)
print(full_path)
```

```python
path='example1.txt'
if os.path.exists(path):
    print(f"The path '{path}' exist.")
else:
    print(f"The path '{path}' doesn't exist.")    
```

```python
# Cheaking is the path is a file or drictory.

path = 'example.txt'
if os.path.isfile(path):
    print(f"Path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"Path '{path}' is a drictory.")
else:
    print(f"path '{path}' is nither file or a drictory.")    
```

```python
## Getting the absolute path

rel_path='example.txt'
abs_path=os.path.abspath(rel_path)
print(abs_path)
```
