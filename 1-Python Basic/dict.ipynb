```python
## Dictonary are mutable.
student={"name":"Nishit", "age":29, "grade":"A"}
print(student)
student["age"]=29 ## Updated value
print(student)
student['address']="India" ## added value
print(student)
del student['grade']
print(student)
```

```python
student={
    "student1":{'name':"Nishit", 'age':28},
    "student2":{'name':'Nalin','age': 29}
}
print(student['student1']['name'])
print(student['student2']['name'])
```

```python
for student_id,student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
squares={x:x**2 for x in range(10)}
print(squares)
```

```python
even_squares={x:x**2 for x in range(10) if x%2==0}
print(even_squares)
```

```python
numbers=[1,2,2,3,3,3,4,4,4,4]

frequency={}

for i in numbers:
    frequency[i]= frequency.get(i, 0)+1
print(frequency)
```
