```python
## Accessing List elements

fruits=["apple", "banana","watermelon", "cherry", "kiwi", "gwava","orange"]
```

```python
print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[-1])
```

```python
print(fruits[1:])
print(fruits[1:3])
```

```python
fruits.append("orrange")
print(fruits)
```

```python
fruits.insert(1, "banana")
print(fruits)
```

```python
fruits.remove("banana")
print(fruits)
```

```python
popped_fruits=fruits.pop()
print(popped_fruits)
print(fruits)
```

```python
index = fruits.index("cherry")
print(index)
```

```python
fruits.sort()
print(fruits)
```
