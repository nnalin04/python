```python
corpus="""Hello woelcome, to udimie's to learn Python.
Please do watch the entire course! to become and expert in NLP.
"""

print(corpus)
```

```python
import nltk
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
from nltk.data import find

print(find("tokenizers/punkt"))
print(find("tokenizers/punkt_tab"))
```

```python
import nltk
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize

sentences = sent_tokenize(corpus)
print(sentences)
```

```python
print(type(sentences))
```

```python
for sentense in sentences:
    print(word_tokenize(sentense))
```

```python


words=word_tokenize(corpus)
print(words)
```

```python
for word in words:
    print(word)
```

```python
from nltk.tokenize import wordpunct_tokenize

wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```
