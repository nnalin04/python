```python
## try, except, else

try:
    num=int(input('Enter a number'))
    result=10/num
except ValueError:
    print(f"This is not a valid number")
except ZeroDivisionError:
    print("Enter a no grater than 0")
except Exception as ex:
    print(ex)
else:
    print(f"The result is '{result}'.")        
```

```python
## try, except, else & finally

try:
    num=int(input('Enter a number'))
    result=10/num
except ValueError:
    print(f"This is not a valid number")
except ZeroDivisionError:
    print("Enter a no grater than 0")
except Exception as ex:
    print(ex)
else:
    print(f"The result is '{result}'.")   
finally:
    print("Execution compleated.")    
```
