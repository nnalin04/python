```python
## Define a class with a instence method

class Dog:
    def __init__(self,name, age):
        self.name=name
        self.age=age

    def bark(self):
        print(f"{self.name} says woof")

dog1=Dog("Buddy", 3)
dog1.bark()
dog2=Dog("lucy", 4)
dog2.bark()
```

```python

## Modeling a Bank Account

## Define a class for Bank Account

class BankAccount:
    def __init__(self, owner, balence=0):
        self.owner=owner
        self.balence=balence

    def deposit(self, amount):
        self.balence+=amount
        print(F"{amount} is deposited. New Balence is {self.balence}")

    def withdraw(self, amount):
        if amount > self.balence:
            print("iinsufficent balence")
        else:
            self.balence-=amount
            print(f"{amount} is withdrawn, new Balence amount is {self.balence}")
```

```python
## Create a new account

account=BankAccount("Nishit", 250000)

print(account.balence)
```

```python
account.deposit(20000)
```

```python
account.withdraw(8000)
```
