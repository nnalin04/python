```python
## Basic Magic Method
class Person():
    def __init__(self, name, age):
        self.name=name
        self.age=age
    
    def __str__(self):
        return f"{self.name}, {self.age} yesrs old"


person=Person("Nishit", 28)    
print(person)    
```
