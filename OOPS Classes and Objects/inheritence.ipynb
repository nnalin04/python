```python
## Inhereritence

class Car:
    def __init__(self, window, door, engine_type):
        self.window=window
        self.door=door
        self.engine_type=engine_type

    def drive(self):
        print(f"The person will drive the S58 {self.engine_type} engine BMW.")    
```

```python
car1=Car(4,6,"petrol")
car1.drive()
```

```python
class Tesla(Car):
    def __init__(self, window, door, engine_type, is_selfdriving):
        super().__init__(window, door, engine_type)
        self.is_selfdriving=is_selfdriving

    def self_driving(self):
        print(f"Tesla support self driving {self.is_selfdriving}.")    
```

```python
tesla1=Tesla(4,6, "electric", True)
```

```python
tesla1.self_driving()
```

```python
### Multiple Inheritence
## when a class inherite from more than one base class


class Animal:
    def __init__(self, name):
        self.name=name

    def speak(self):
        print("Sub Class must implement this method")

class Pet:
    def __init__(self, owner):
        self.owner=owner            
```

```python
class Dog(Animal,Pet):
    def __init__(self, name, owner):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)

    def speak(self):
        return f"{self.name} says woof"    
```

```python
dog=Dog("Buddy", "Nishit")

print(dog.speak())
print(f"Owner: {dog.owner}")
```
